```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dd=pd.DataFrame(np.random.randint(1,100,(5,3)),index=["mahir","heet","yash","meet","mihir"],columns=["python","datascience","DBMS"])
dd
```

```python
plt.plot(dd.index,dd.python,"g--s",label="python",lw=0.25,marker=("d"),ls=(":"))
#style to line("-","--",":","-.")
#colour to line(rgbcymk)
#marker (o,.,,,x,v,^,<,>,s,d)
plt.plot(dd.index,dd.datascience,"r--s",label="datascience")
plt.title("student result")
plt.xlabel("Name")
plt.ylabel("python marks")
plt.legend()
plt.ylim(-20,100)
```

```python
plt.subplot(2,2,1)
plt.plot(dd.index,dd.python,"b--s",label="python",lw=0.25,marker="o")
plt.ylabel("python marks")
plt.title("python")
plt.subplot(2,2,2)
plt.plot(dd.index,dd.datascience,"r--",label="datascience",marker="o")
plt.ylabel("datascience marks")
plt.title("datascience")
plt.subplot(2,2,3)
plt.plot(dd.index,dd.DBMS,"c-.",label="DBMS",marker="d")
plt.ylabel("DBMS marks")
plt.title("DBMS")
plt.subplot(2,2,4)
plt.plot(dd.index,dd.python,"m--s",label="python",lw=1.25,marker="^")
plt.plot(dd.index,dd.datascience,"m--",label="datascience",lw=1.25,marker="^")
plt.plot(dd.index,dd.DBMS,"m--",label="DBMS",lw=1.25,marker="^")
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,6,11)
y=x**2
#create Figure
fig=plt.figure()
#Add set of axes to figure
axes=fig.add_axes([0.01,0.01,0.3,0.3])#left,bottom,width,hight
#plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('set x label')#notice the use of set_ to begin methods
axes.set_ylabel('set y label')
axes.set_title('set Title')
```

```python
x=np.linspace(0,6,11)
y=x**2
#create Figure
fig=plt.figure()
#Add set of axes to figure
axes=fig.add_axes([0.01,0.01,0.6,0.6])#left,bottom,width,hight
axes1=fig.add_axes([0.10,0.3,0.2,0.2])
#plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('set x label')#notice the use of set_ to begin methods
axes.set_ylabel('set y label')
axes.set_title('set Title')
axes1.plot(x,y,'r')
axes1.set_xlabel('set x label')#notice the use of set_ to begin methods
axes1.set_ylabel('set y label')
axes1.set_title('set Title')
```

```python
fig1,ax=plt.subplots(2,3,sharex=True,sharey=True,layout="constrained")
ax[1,2].plot(x,y)
ax[1,2].set_title("last")
ax[1,2].set_ylabel("y axis")
ax[0,1].plot(x,x)
```

```python
dd=pd.DataFrame(np.random.randint(1,100),index=["mahir","heet","yash","meet","mihir"],columns=["python","datascience","DBMS","DSA"])
dd
fig1,ax=plt.subplots(2,2,sharex=True,sharey=True,layout="constrained")
ax[0,0].plot(dd.index,dd.python,"r:",label="python")
ax[0,0].set_title("python")
ax[0,0].legend(loc=1)
ax[0,1].plot(dd.index,dd.datascience,"g^-",label="datascience")
ax[0,1].set_title("datascience")
ax[0,1].legend(loc=1)
ax[1,0].plot(dd.index,dd.DBMS,"ks-",label="DBMS")
ax[1,0].set_title("DBMS")
ax[1,0].legend(loc=1)
ax[1,1].plot(dd.index,dd.DSA,"bo-",label="DSA")
ax[1,1].set_title("DSA")
ax[1,1].legend(loc=1)
```

Seaborn 
=>distplot=univariant data,density+histogram with kernel density estimation plot(kde)plot.
=>jointplot=visualise the joint distribution of two variables along with their individual distribution
=>pairplot=useful for exploring relationships in multivariate dataset,shows pairwies relationship between variable in dataset
=>rugplot=display individual data point along a single axes
=>kdeplot=estimates probability density function of a continous random variable
